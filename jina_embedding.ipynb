```python
from transformers import AutoModel
from numpy.linalg import norm
```

```python
cos_sim = lambda a,b: (a @ b.T) / (norm(a)*norm(b))
model = AutoModel.from_pretrained('jinaai/jina-embeddings-v2-base-en', trust_remote_code=True) # trust_remote_code is needed to use the encode method
```

```python
sentence_list=['How is the weather today?', 'What is the current weather like today?']

embeddings = model.encode(sentence_list)
```

```python
print(cos_sim(embeddings[0], embeddings[1]))
```

```python
!rm -rf /content/sample_data/vector_storage_demo
!mkdir /content/sample_data/vector_storage_demo

storage_path="/content/sample_data/vector_storage_demo/chroma_db"

import chromadb
from chromadb.config import Settings


client = chromadb.Client(Settings(
                                    persist_directory=storage_path
                                ))

collection = client.get_or_create_collection(name="Studentv")


sentence_list=[student_info, club_info, university_info]
students_embeddings = model.encode(sentence_list)

collection.add(
    embeddings = students_embeddings.tolist(),
    documents = [student_info, club_info, university_info],
    metadatas = [{"source": "student info"},{"source": "club info"},{'source':'university info'}],
    ids = ["id1", "id2", "id3"]
)
```

```python
!ls /content/sample_data/vector_storage_demo
```

```python
query="Who is Alexandra?"

results = collection.query(
    query_embeddings=[model.encode(query).tolist()],
    n_results=1
)

print(results)
```
